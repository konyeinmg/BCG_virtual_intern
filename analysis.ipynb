```python
churn = client_data[['id', 'churn']]
churn.columns = ['Companies', 'churn']
churn.head()
```

```python
churn_total = churn.groupby('churn').count()
churn_total
```

```python
churn_percentage = (churn_total / churn_total.sum()) * 100
churn_percentage
```

```python
churn_percentage.transpose()
```

```python
def annotate_stacked_bars(ax, pad=0.99, colour="white", textsize=13):
    """
    Add value annotations to the bars
    """

    # Iterate over the plotted rectanges/bars
    for p in ax.patches:
        
        # Calculate annotation
        value = str(round(p.get_height(),1))
        # If value is 0 do not annotate
        if value == '0.0':
            continue
        ax.annotate(
            value,
            ((p.get_x()+ p.get_width()/2)*pad-0.05, (p.get_y()+p.get_height()/2)*pad),
            color=colour,
            size=textsize
        )
```

```python
import matplotlib.pyplot as plt
```

```python
ax = churn_percentage.transpose().plot(kind = 'bar', stacked = True, figsize = (8,6), rot = 0)
annotate_stacked_bars(ax)
plt.legend(['Retention', 'Churn'], loc='upper right')
plt.ylabel('Company base (%)')
```

```python
channel = client_data[['id', 'channel_sales', 'churn']]
channel = channel.groupby([channel['channel_sales'], channel['churn']])['id'].count().unstack(level=1)
```

```python
channel
```

```python
channel_churn = (channel.div(channel.sum(axis=1), axis = 0) * 100).sort_values(by=[1], ascending = False)
channel_churn
```

```python
ax = channel_churn.plot(kind = 'bar', stacked = True, figsize = (8,6), rot = 75)
annotate_stacked_bars(ax, textsize = 6)
plt.legend(['Retention','Churn'],loc='upper right')
plt.ylabel('Company base(%)')
```

```python
consumption = client_data[["cons_12m", "cons_gas_12m", "cons_last_month", "imp_cons"]]
consumption.head()
```

```python
for col in consumption:
    ax = client_data.boxplot(column = col, by = 'churn', figsize = (6,6))
    plt.title(col)
plt.show()
```

```python
def plot_distribution(dataframe, column, ax, bins=50):
    temp = pd.DataFrame({'Rentention' : client_data[client_data['churn'] == 0][column], 'Churn' : client_data[client_data['churn'] == 1][column]})
    temp.plot(kind = 'hist', bins = bins, ax = ax, stacked = True)
    ax.set_xlabel(column)
    ax.ticklabel_format(style='plain', axis='x')
```

```python
consumption = client_data[["cons_12m", "cons_gas_12m", "cons_last_month", "imp_cons", "has_gas", "churn"]]
```

```python
fig, axs = plt.subplots(nrows=4, figsize=(18, 25))

plot_distribution(consumption, 'cons_12m', axs[0])
plot_distribution(consumption[consumption['has_gas'] == 't'], 'cons_gas_12m', axs[1])
plot_distribution(consumption, 'cons_last_month', axs[2])
plot_distribution(consumption, 'imp_cons', axs[3])
```
